```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
# Watermark for one subplot in a figure
fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(221)
ax1.plot([1, 2, 3, 4], [2, 4, 6, 8])
ax1.set_label('straight line')
ax1.text(1, 4, 'Do not distribute',
         fontsize=20,
         color='red',
         ha='left',
         va='bottom',
         alpha=0.5)
```

```python
fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(221)
ax1.plot([1, 2, 3, 4], [2, 4, 6, 8])
ax1.set_label('straight line')
ax1.text(1, 4, 'Do not distribute',
         fontsize=20,
         color='red',
         ha='left',
         va='bottom',
         alpha=0.5)

ax2 = fig.add_subplot(222)
ax2.plot(x, np.sin(x))

ax3 = fig.add_subplot(223)
ax3.plot(x, np.cos(x))
```

```python
"""
    A watermark for the entire figure in addition to the one we have for one subplot
    
"""

fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(221)
ax1.plot([1, 2, 3, 4], [2, 4, 6, 8])
ax1.set_label('straight line')
ax1.text(1, 4, 'Do not distribute',
         fontsize=20,
         color='red',
         ha='left',
         va='bottom',
         alpha=0.5)

ax2 = fig.add_subplot(222)
ax2.plot(x, np.sin(x))

ax3 = fig.add_subplot(223)
ax3.plot(x, np.cos(x))

fig.text(0.1, 0.5, 'Property of chepkoy',
        fontsize=40,
        color='gray',
        ha='left',
        va='bottom',
        alpha=0.3)
```
