```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.interactive(True)
plt.ion()
mpl.is_interactive()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
"""
    Arguments: 1st -> This represents where in the figure the left of the axes will lie. 0 is the extreme left 1 is extreem right
               2nd -> This represents the bottom location of the axes in the figure 0 is the bottom of the figure 1 is the top
               3rd -> width of the axes in terms of the fraction of the figure width
               4th -> Height of the axes in terms of the fraction of the figure height
"""
plt.show()
```

```python
type(ax)
```

```python
"""
    Plotting a figure with 2 axes: Ideal for side by side comparison

"""
x = np.linspace(start=0, stop=10, num=50)
fig=plt.figure()
ax1 = fig.add_axes([0, 0.6, 1, 0.4])
ax2 = fig.add_axes([0, 0, 0.8, 0.4])

ax1.plot(x, np.sin(x), color='m')
ax1.set_xlabel('x', fontsize=15, color='m')
ax1.set_ylabel('sin(x)', fontsize=15, color='m')

ax2.plot(x, np.cos(x), color='red')
ax2.set_xlabel('x', fontsize=15, color='red')
ax2.set_ylabel('cos(x)', fontsize=15, color='red')

plt.show()
```

```python
# Overlapping axes
x = np.linspace(start=0, stop=10, num=50)
fig=plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.4])
plt.show()
```

```python
# Explicitly set the figure width and height in Inches
fig = plt.figure(figsize=(8,8))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.4])

plt.show()
```

```python
"""
    Another way of setting up mulptiple axes within a matplotlib figure layout in a matrix format is to use
    the subplot functionality
    function: subplot() -> will add a set of axes in a 2x2 grid to the 1st position (top left)
    
    Argument: e.g 221 -> 2x2 grid to the 1st position in the grid 1st cell top left
    
    Accepts only valid values -> 5 will give a value error
"""

fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(221)
plt.show()
```

```python
# compare with a matplotlib.axes._axes.Axes object returned by fig.add_axes()
type(ax1)
```

```python
# An AxesSubplot is a type of Axes
isinstance(ax1, mpl.axes._axes.Axes)
```

```python
fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(221)
ax1.plot([1, 2, 3, 4],[2, 4, 6, 8])

# To the same figure as the previous one -plus a new subplot with sine curve
ax2 = fig.add_subplot(222)
ax2.plot(x, np.sin(x))

# Now with a 3rd set of axes
ax3 = fig.add_subplot(223)
ax3.plot(x, np.cos(x))
plt.show()
```

```python
"""
    The subplot2grid() function is another way of specifying subplots
    arguments: 2 tuples
    1. Matrix representation of the grid
    2. Row /colum values
    3. addittional feature you can have subplots spanmultiple grid positions either col or row
    
"""

ax1 = plt.subplot2grid((2, 3), (0, 0))
ax1.plot(x, np.sin(x))
ax1.set_label('sine curve')

ax2 = plt.subplot2grid((2, 3), (0, 1))
ax2.plot(x, np.cos(x))
ax2.set_label('cos curve')

ax3 = plt.subplot2grid((2, 3), (0, 2), rowspan=2)
ax3.plot([1, 2, 3, 4], [2, 4, 6, 8])
ax3.set_label('straight line')
ax3.yaxis.tick_right()

ax4 = plt.subplot2grid((2, 3), (1, 0), colspan=2)
ax4.plot(x, np.exp2(x))
ax4.set_label('exponential curve')

```
