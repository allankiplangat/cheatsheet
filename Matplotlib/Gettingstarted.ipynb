```python
x = np.linspace(0, 10, 50)
```

```python
x
```

```python
plt.title('Plots getting started')
plt.plot(x, np.sin(x), label='sine curve')
plt.legend()

plt.xlabel('x', fontsize=15, color="blue")
plt.ylabel('sin(x)', fontsize=15, color="red")
plt.tick_params(
    axis='y',
    color="red",
    labelcolor='blue',
    labelsize='xx-large'
)
plt.tick_params(
    axis='x',
    bottom=False,
    labelbottom=False
)
```
