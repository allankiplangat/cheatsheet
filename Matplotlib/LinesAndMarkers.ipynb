```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.interactive(True)
plt.ion()
mpl.is_interactive()
```

```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
x
```

```python
# Adding multiple curves
plt.plot(x, np.sin(x), label='sine curve', color='green')
plt.plot(x, np.cos(x), label='cosine curve', color='m')
plt.legend()
plt.title("Adding multiple plots")
plt.show()
```

```python
random_array = np.random.rand(20)
plt.plot(
    random_array, 
    color='green',
    linestyle=':', # linestyle='None' to get rid of the line
    #linestyle='-',
    linewidth=1,
    marker = 'd', # use diamond-shaped markers to denote the points
    markersize=6,
)
plt.show()
```

```python
# Scatter Plot
plt.scatter(
    range(0,20), #x
    random_array,# y
    color='green',
    marker='d')
plt.show()
```
